```python
categorical_labels = [['YearsOfOnlineShopping', 'PurchaseInPast1Year'], ['InternetAccess', 'DeviceUsed']]
colors = [['Set1', 'Set2'], ['Set3', 'PuRd']]

fig, ax = plt.subplots(2, 2, figsize=(18, 12))
for i in range(2):
    for j in range(2):
        feature = categorical_labels[i][j]
        color = colors[i][j]
        ax1 = sns.countplot(x=feature,  palette=color, data=df1, ax=ax[i][j])
        ax1.set_xlabel(feature, labelpad=10)
       
sns.despine()
```

```python

categorical_feats = df1.dtypes[df1.dtypes == "object"].index
print("Number of Categorical features: ", len(categorical_feats))
```

```python
li_cat_feats = list(categorical_feats)
nr_rows = 35
nr_cols = 2

fig, axs = plt.subplots(nr_rows, nr_cols, figsize=(nr_cols*8,nr_rows*10))

for r in range(0,nr_rows):
    for c in range(0,nr_cols):  
        i = r*nr_cols+c
        if i < len(li_cat_feats):
            ax1=sns.countplot(x=li_cat_feats[i], data=df1, ax = axs[r][c])
            ax1.set_xlabel(li_cat_feats[i], labelpad=10)
    
sns.despine()  
```
